Write accented files

```python
counter = 1

for path, y, files in os.walk('to_acc/'):
  for f in files:
    if f.endswith(".txt"):
        try:
            accent.write_file([os.path.join(path, f)])
            print(counter, f)
            counter += 1
        except:
            print(counter, "Problem occured with:", f)
            counter += 1
```
